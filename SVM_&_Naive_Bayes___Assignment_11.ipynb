### Ans:1-

Support Vector Machine (SVM) is a machine learning algorithm used for:

   *  Classification (e.g., spam vs not spam)    
   *  Regression (predicting continuous values)

It is mainly used for classification tasks.

SVM tries to find the best line (or boundary) that separates different classes of data. This line is called the hyperplane.

 # how it works:-
Imagine you have two types of points: Positive(+), Negative(-).

SVM will:
   * Find a line (hyperplane) that separates them.

   * Maximize the distance between the line and the closest points from each class.

       These closest points are called Support Vectors.

       The bigger the gap (margin), the better the separation.

# Ans:-2

   ### Hard Margin SVM?
  * It does not allow any mistakes.

  * All data points must be perfectly separated by the hyperplane.

  * Works only when the data is linearly separable (no overlap between classes)
  
### Soft Margin SVM?
  * It allows some mistakes (a few points can be on the wrong side).

  * Adds flexibility to handle overlapping or noisy data.

  * Tries to balance:

     * Having a wide margin

     * Making fewer classification errors

#Ans:-3

  ### Kernel Trick?
The Kernel Trick is a smart mathematical method used in SVM to handle data that you can’t split it with a straight line.

It transforms the data into a higher-dimensional space where a straight line can separate the classes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
Digit = datasets.load_digits()
Digit
```

```python
print(Digit.DESCR)
```

```python
x= Digit.data
y= Digit.target
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6, 4))
for index, image in enumerate(Digit.images[:6]):
    plt.subplot(2, 3, index + 1)
    plt.imshow(image, cmap='gray')
    plt.title(f"Label: {Digit.target[index]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.svm import SVC
model= SVC(kernel='rbf', gamma='scale', C=1.0)
model.fit(x_train, y_train)
```

```python
y_prd= model.predict(x_test)
```

```python
y_prd
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_test, y_prd))
```

```python
print(confusion_matrix(y_test, y_prd))
```

```python
print(classification_report(y_test, y_prd))
```

```python
conf_matrix = confusion_matrix(y_test, y_prd)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=Digit.target_names, yticklabels=Digit.target_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## Ans:-4

  ### Naïve Bayes Classifier?
Naïve Bayes is a machine learning algorithm used mainly for classification tasks.
It is based on Bayes Theorem, which helps calculate the probability of something happening, based on prior knowledge.

  * It is especially popular for tasks like:

    * Spam detection

    * Sentiment analysis

    * Text classification

  ### Why is it called “Naïve”?
It’s called “naïve” because it assumes that:

All features (input variables) are independent of each other, which is often not true in real life.

This assumption is too simple (naïve), but surprisingly, the model still works very well for many problems, especially with text data.

## Bayes' Theorem Formula:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$


* **P(A|B)** = Probability of A given B
* **P(B|A)** = Probability of B given A
* **P(A)** = Probability of A
* **P(B)** = Probability of B

# Ans:-5

## 1. Gaussian Naïve Bayes

### Used For:

* When the features are continuous numeric values
* Example: Height, weight, temperature, etc.

### How it works:

* Assumes the data follows a Normal (Gaussian) distribution
* It uses mean and standard deviation to calculate probabilities

### Use Case Example:

* Medical diagnosis based on patient measurements (e.g., blood pressure, age)
* Iris flower classification using petal/sepal measurements

##  2. Multinomial Naïve Bayes

### Used For:

* When the data is in the form of counts or frequencies
* Most commonly used for text classification

### How it works:

* Counts how often each feature (e.g., word) appears in a document
* Doesn’t work well with negative or continuous data

### Use Case Example:

* Spam vs. non-spam email classification
* News article topic classification (sports, politics, etc.)
* Sentiment analysis using word counts

##  3. Bernoulli Naïve Bayes

### Used For:

* When features are binary (yes/no, 0/1, true/false)
* It checks whether a feature is present or absent

### How it works:

* Only cares if a word appears or not, not how many times

### Use Case Example:

* Spam detection: does the email contain the word "free"? (1 or 0)
* Document classification where you care about presence not frequency

```python
from sklearn.datasets import fetch_20newsgroups
categories = ['sci.med', 'comp.graphics']
data3 = fetch_20newsgroups(subset='all', categories=categories, remove=('headers', 'footers', 'quotes'))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
x = vectorizer.fit_transform(data3.data)
y = data3.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
x_train.shape, x_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred= model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_test, y_pred)
```

```python
print(roc_auc)
```
