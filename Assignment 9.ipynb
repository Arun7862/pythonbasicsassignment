#   Statistics Part 2 Theory


Ans-1:-   Hypothesis testing in statistics is a method used to evaluate a claim about a population parameter by analyzing sample data.

Ans-2:-   The null hypothesis (H0) assumes no significant difference or relationship between variables, while the alternative hypothesis (Ha or H1) proposes that a significant difference or relationship exists.

Ans-3:-   In hypothesis testing, the significance level (α) is the probability of rejecting the null hypothesis when it is actually true.

Ans-4:-   A p-value represents the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true.

Ans-5:-   The p-value represents the probability of obtaining results as extreme as, or more extreme than, the one calculated from your sample data, assuming the null hypothesis is true.

Ans-6:-   In hypothesis testing, Type I and Type II errors are the two types of mistakes that can occur when drawing conclusions about a population based on sample data.

Ans-7:-   Type I error occurs when the null hypothesis is wrongly rejected, indicating a false positive. A Type II error occurs when the null hypothesis is wrongly retained, indicating a false negative.

Ans-8:-   A Z-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two independent samples, when the population variance is known and the sample size is large (n ≥ 30).

Ans-9:-   The Z-score measures how many standard deviations a data point is away from the mean of a dataset.
represent:- Z = (x - μ) / σ

Ans-10:-   The t-distribution is a probability distribution similar to the normal distribution but with heavier tails. It's particularly useful when dealing with small sample sizes or when the population standard deviation is unknown.

Ans-11:-   Z-tests are used when the sample size is large (generally > 30) or the population standard deviation is known.
T-tests are used when the sample size is small (generally < 30) and the population standard deviation is unknown.

Ans-12:-   A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups.

Ans-13:-   Z-tests are typically used when the population standard deviation is known and the sample size is large (usually > 30), while t-tests are used when the population standard deviation is unknown or the sample size is small (usually < 30).

Ans-14:-   A confidence interval (CI) is a range of values that is likely to contain the true value of an unknown population parameter, such as the population mean.

Ans-15:-   The margin of error is a measure of the potential difference between a sample result and the true population result.

Ans-16:-   Bayes Theorem is a fundamental tool in statistics for updating beliefs or probabilities based on new evidence.

Ans-17:-   The chi-square distribution is a continuous probability distribution often used in hypothesis testing, particularly when analyzing categorical data or comparing observed frequencies to expected frequencies.

Ans-18:-   step:-1. Define the hypothesis
           step:-2. Calculate expected frequencies
           step:-3. Calculate the chi-square test statistic
           step:-4. Determine degrees of freedom
           step:-5. Find the p-value
           step:-6. Make a decision

Ans-19:-   The F-distribution is a continuous probability distribution used primarily in hypothesis testing, particularly when comparing variances or means across multiple groups.

Ans-20:-   An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to determine if there's a statistically significant difference between them.

Ans-21:-   There are several types of ANOVA, including one-way, two-way, and factorial ANOVA.

Ans-22:-   The F-test is a statistical test that compares variances between two or more groups or samples. The F-test, often used in ANOVA calculates an F-statistic and compares it to a critical value from the F-distribution.

```python
#Ans-9:-
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_and_plot_zscores(data):
  z_scores = stats.zscore(data)

  plt.figure(figsize=(10, 6))
  plt.scatter(range(len(z_scores)), z_scores, marker='o', linestyle='-')
  plt.title('Z-Scores Plot')
  plt.xlabel('Data Point Index')
  plt.ylabel('Z-Score')
  plt.grid(True)
  plt.axhline(0, color='black', linewidth=0.8)
  plt.show()
```

```python
#Ans:1
import numpy as np
from scipy.stats import norm

population_mean = 50
population_std = 8

sample = [52, 49, 51, 50, 53, 48, 47, 54, 49, 52]
sample_mean = np.mean(sample)
sample_size = len(sample)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
#Ans:6
import numpy as np
from scipy import stats

def calculate_confidence_interval(sample_mean, sample_std, sample_size, confidence_level):
    alpha = 1 - confidence_level
    margin_of_error = stats.t.ppf(1 - alpha / 2, df=sample_size - 1) * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound
```

```python
#Ans:7
import numpy as np
from scipy import stats

before_intervention = np.array([75, 80, 85, 78, 90, 82, 79, 88, 85, 92])
after_intervention = np.array([78, 85, 87, 80, 92, 84, 81, 90, 88, 95])

t_statistic, p_value = stats.ttest_rel(before_intervention, after_intervention)

print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
```
